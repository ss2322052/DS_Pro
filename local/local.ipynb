```python
import sqlite3
import pandas as pd

# CSVファイルのパス
file_path = '/Users/hayakawafutaba/Lecture/DSPR/DS_Pro/local/local.csv'  

# CSVファイルの読み込み
df = pd.read_csv(file_path)

# SQLiteデータベースの作成（または接続）
conn = sqlite3.connect('local.db')  

# カーソルの作成
cursor = conn.cursor()

# テーブルの作成
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sleep_data (
        date TEXT,
        sleep_hours INTEGER
    )
''')

# CSVからデータを読み込み、データベースに挿入
df.to_sql('sleep_data', conn, if_exists='replace', index=False)

# 挿入されたデータの確認
cursor.execute('SELECT * FROM sleep_data')
inserted_data = cursor.fetchall()
for row in inserted_data:
    print(row)

# データベースとの接続を閉じる
conn.close()
```
